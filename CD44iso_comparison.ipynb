```python
import pandas as pd
cl =  'FPKM.tsv'
an = 'CD44_isoforms.tsv'
samples = pd.read_csv(cl, sep='\t')
iso = pd.read_csv(an, sep='\t')
```

```python
# CD44_id
CD44_id = iso.uc001mvu.unique()
iso = iso.rename(columns={'CD44-208': 'names'})
CD44_names = iso.names.unique()
CD44_names
```

```python
samples.index.name = 'index'
samples
```

```python
samples = samples.reset_index()
CD44_samples = samples[(samples['index'].isin(['uc001mvu', 'uc001mvv', 'uc001mvw', 'uc001mvx', 
                                               'uc001mvy', 'uc001mwc', 'uc010rer']))]
CD44_samples = CD44_samples.set_index('index')
CD44_samples
```

```python
CD44_samples.to_csv('CD44_samples.csv', index=False)
# выбрала через терминал колонки, где в середине '01, 02, 06' и '11'
```

```python
ca = 'CD44_samples_cancer.tsv'
he = 'CD44_samples_healthy.tsv'
cancer_samples = pd.read_csv(ca, sep='\t')
healthy_samples = pd.read_csv(he, sep='\t')
cancer_samples = cancer_samples.set_index('index')
healthy_samples = healthy_samples.set_index('index')
healthy_samples
```

```python
isoforms_c = ['CD44-208', 'CD44-206', 'CD44-209', 'CD44-201', 'CD44-203', 'CD44-210','CD44-205']*287
isoforms_h = ['CD44-208', 'CD44-206', 'CD44-209', 'CD44-201', 'CD44-203', 'CD44-210','CD44-205']*41
```

```python
cancer_s = pd.melt(cancer_samples)
healthy_s = pd.melt(healthy_samples)
cancer_s
```

```python
type_h = ['healthy']*287
type_c = ['cancer']*2009
cancer_s['Iso'] = isoforms_c
healthy_s['Iso'] = isoforms_h
cancer_s['type'] = type_c
healthy_s['type'] = type_h
cancer_s = cancer_s.rename(columns={"variable": "TCGA", "value": "Expression"})
healthy_s = healthy_s.rename(columns={"variable": "TCGA", "value": "Expression"})
all_samples = healthy_s.append(cancer_s)
```

```python
import seaborn as sns
import matplotlib as plt
plot = sns.boxplot(y='Expression', x='Iso', 
                 data=all_samples, 
                 width=0.5,
                 palette="Set2", hue='type')

plot.set_xticklabels(labels=['CD44-208', 'CD44-206', 'CD44-209', 'CD44-201', 'CD44-203', 'CD44-210','CD44-205'], rotation=30)
plot.get_figure().savefig('CD44_iso_expr.png', dpi=300)
```

```python
import seaborn as sns
plot = sns.boxplot(y='Expression', x='Iso', 
                 data=healthy_s, 
                 width=0.5,
                 palette="Set2")
```

```python
import matplotlib.pyplot as plt
numbers_c = [1, 2, 3, 4, 5, 6, 7]*287
numbers_h = [1, 2, 3, 4, 5, 6, 7]*41
cancer_s['Isoforms'] = numbers_c
healthy_s['Isoforms'] = numbers_h
gr1 = sns.catplot(x="Isoforms", y="Expression", hue="Iso",
                data=healthy_s, palette='Set2', legend_out=True)
gr1.fig.suptitle('Healthy samples')
gr1.savefig('healthy_expr.png')
```

```python
gr2 = sns.catplot(x="Isoforms", y="Expression", hue="Iso",
                data=cancer_s, palette='deep', legend_out=True)
gr1.fig.suptitle('Cancer samples')
gr2.savefig('cancer_expr.png')
```
