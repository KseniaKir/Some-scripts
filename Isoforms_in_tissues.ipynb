```python
import pandas as pd
import numpy as np

tsv =  'CCLE_RNAseq_rsem_transcripts_tpm_20180929.tsv'
df = pd.read_csv(tsv, sep='\t', index_col=1)
df = df.drop(df.columns[0], axis=1)

df[1:1019] = df[1:1019].apply(pd.to_numeric)
df = df.add(1)
df = df.apply(np.log2)
df.head()
```

```python
tissues = ['HT29', 'HCT116', 'MEWO', 'MV', 'PACA', 'BXPC', 'L3', 
           'HOS', 'SKOV3', 'MOLM','HMC', 'LG']
# [col for col in df.columns if 'PACA' in col]
```

```python
for i in tissues:
    print('GENE', i, 'isin', [col for col in df.columns if i in col])
    print(' ')
```

```python
selected = ['SKOV3_OVARY', 'BXPC3_PANCREAS', 'MEWO_SKIN', 'HCT116_LARGE_INTESTINE', 'HT29_LARGE_INTESTINE', 'MV411_HAEMATOPOIETIC_AND_LYMPHOID_TISSUE', 'L33_PANCREAS', 'L363_HAEMATOPOIETIC_AND_LYMPHOID_TISSUE',
            'HOS_BONE', 'MOLM13_HAEMATOPOIETIC_AND_LYMPHOID_TISSUE', 'MOLM16_HAEMATOPOIETIC_AND_LYMPHOID_TISSUE',
            'MOLM6_HAEMATOPOIETIC_AND_LYMPHOID_TISSUE', 'HMC18_BREAST']
```

```python
fin = []
for i in selected:
    m = pd.Series(df.loc[:, i])
    fin.append(m)
#df_sel = df_sel.set_index('ENST')
df_sel = pd.DataFrame(fin, columns=df.index)
df_sel.columns = df_sel.columns.map(lambda x: str(x)[:-2])
df_sel
```

```python
cl = 'CD44_isoforms.tsv'
df_ann = pd.read_csv(cl, sep='\t', index_col=1)
ENST_CD44 = df_ann.index.to_list()
ENST_CD44.append('ENST00000428726')
df_ann
```

```python
l3 = df_ann.iloc[:, 3].to_list()
l3.append('CD44-208')
```

```python
CD44_samples = df_sel[ENST_CD44]
j = 0
for i in CD44_samples.columns:
    CD44_samples = CD44_samples.rename(columns={i: l3[j]})
    j+=1
CD44_samples.to_csv('CD44_in_cancer_tissues.csv', index=True)
```

```python
import seaborn as sns
cl = sns.clustermap(CD44_samples, cmap="Blues")
#plt.setp(cl.ax_heatmap.get_xticklabels(), rotation=90)
cl.savefig('cluster.png', dpi=300)
```
